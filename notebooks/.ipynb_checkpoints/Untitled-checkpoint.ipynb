```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import collections
import itertools
```

```python
VIDEO_PATH = "../videos/"
INPUT_FILEPATH = VIDEO_PATH + "RPI Men's Hockey vs. Cornell University.mp4"
OUTPUT_FILEPATH = VIDEO_PATH + "clip1 improved next.mp4"

skip_frame_mod = 3

buffer_top = 150
buffer_bot = 200
buffer_left = 900
buffer_right = 350
scale = .3

longer_frames_length = 10
longer_frames_mod = 10
previous_frames_length = 5

print_status_mod = 10



input_video = cv2.VideoCapture(INPUT_FILEPATH)

width = int(input_video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(input_video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = input_video.get(cv2.CAP_PROP_FPS)


test = np.ones((width, height))
test = cv2.resize(test, (0,0), fx=scale, fy=scale)
fourcc = cv2.VideoWriter_fourcc(*'X264')
output_video = cv2.VideoWriter(OUTPUT_FILEPATH, fourcc, fps/skip_frame_mod, (test.shape[0]+ buffer_right + buffer_left,test.shape[1] + buffer_top + buffer_bot))
```

```python
longer_frame_Ms = collections.deque(maxlen=longer_frames_length)
previous_frame_Ms = collections.deque(maxlen=previous_frames_length)

ret, frame = input_video.read()
previous_frame_Ms.append((frame, np.array([[1,0,0], [0,1,0]])))

currentFrame = 0
while(input_video.isOpened()):
    # Capture frame-by-frame
    ret, frame = input_video.read()

    if ret == True:
        if(currentFrame % skip_frame_mod == 0):
            Ms = list()
            for prev_frame, prev_M in itertools.chain(longer_frame_Ms, previous_frame_Ms):
                M_from_frame = cv2.estimateRigidTransform(frame, prev_frame, False)
                if M_from_frame is not None:
                    
                    angle = np.arctan2(M_from_frame[1][0], M_from_frame[1][1])
                    frame_scale = M_from_frame[1][0]/np.sin(angle)
                    M_from_frame[0:2][0:2] /= frame_scale
                    
                    M = np.matmul(np.concatenate([M_from_frame, np.array([[0,0,1]])]), np.concatenate([prev_M, np.array([[0,0,1]])]))[0:2]
                    Ms.append(M)
            if not Ms:
                print("oops no matches, breaking")
                break
            M = np.mean(Ms, axis = 0)
            #M[0][2] *= scale
            M[1][2] *= scale

            buffered_image = cv2.resize(frame, (0,0), fx=scale, fy=scale)
            buffered_image = cv2.copyMakeBorder(buffered_image, buffer_top, buffer_bot, buffer_left, buffer_right, 0)
            buffered_image = cv2.warpAffine(buffered_image,M,buffered_image.T.shape[1:3])

            previous_frame_Ms.append((frame, M))
            if(currentFrame % longer_frames_mod == 0):
                longer_frame_Ms.append((frame, M))

            output_video.write(buffered_image)
    else:
        print("done")
        break
    
    if(currentFrame % print_status_mod == 0):
        print(currentFrame)

    currentFrame += 1

input_video.release()
output_video.release()
    
```
