```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import collections
import itertools
```

```python
VIDEO_PATH = "../videos/"
INPUT_FILEPATH = VIDEO_PATH + "RPI Men's Hockey vs. Cornell University.mp4"
OUTPUT_FILEPATH = VIDEO_PATH + "clip1 improved next.mp4"
```

```python
input_video = cv2.VideoCapture(INPUT_FILEPATH)

width = int(input_video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(input_video.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = input_video.get(cv2.CAP_PROP_FPS)

scale = .3

skip_frame_mod = 20
```

```python
frames = list()
currentFrame = 0
while(input_video.isOpened()):
    # Capture frame-by-frame
    ret, frame = input_video.read()
    
    if ret == True:
        if currentFrame % skip_frame_mod == 0:
            frames.append(frame)
        
    else:
        print("done")
        break
    
    if(currentFrame % skip_frame_mod == 0):
        print(currentFrame)

    currentFrame += 1

input_video.release()
```

```python
print("creating stitcher")
stitcher = cv2.createStitcherScans()
```

```python
print("[INFO] stitching images...")
new = [frames[i] for i in [5, 4]]
(status, stitched) = stitcher.stitch(new)
```
